```python
import pandas as pd
import requests
import sqlite3
```

```python
conn = sqlite3.connect("database.sqlite")
cur = conn.cursor()
```

```python
cur.execute("""SELECT *
               FROM Matches
               ;""")

matches_df = pd.DataFrame(cur.fetchall())
matches_df.columns = [x[0] for x in cur.description]
matches_df

```

```python
cur.execute("""SELECT *
               FROM Teams_in_Matches""")

match_team_df = pd.DataFrame(cur.fetchall())
match_team_df.columns = [x[0] for x in cur.description]
match_team_df
```

```python
cur.execute("""SELECT *
               FROM Unique_Teams""")

unique_teams_df = pd.DataFrame(cur.fetchall())
unique_teams_df.columns = [x[0] for x in cur.description]
unique_teams_df
```

```python
weather_df = match_team_df.merge(unique_teams_df, how="left", on="Unique_Team_ID")
weather_df
weather_df.merge(matches_df[["Match_ID", "Div", "Date"]], how="left", on="Match_ID") 
```

### Looking for a way to create a column attached to the matches_df that says the team name if they won, then using that column with the match id to join to the weather_df to create another column that says win or lose

```python
practice_df = matches_df

practice_df['Winner'] = "to fill"
for i, result in enumerate(practice_df['FTR']):
    if result == "H":
        practice_df.loc[i, 'Winner'] = practice_df.loc[i, 'HomeTeam']
        
    if result == "A":
        practice_df.loc[i, 'Winner'] = practice_df.loc[i, 'AwayTeam']
        
    if result == "D":
        practice_df.loc[i, 'Winner'] = "draw"
        
practice_df
```
